```julia
using PyPlot
```

```julia
Y1 = [1, 7, 11, 13, 15, 16]
Y2 = [15, 3, 13, 2, 7, 1]

fig, ax = subplots()

ax.plot(Y1)
ax.plot(Y2)

ax.set_xlabel("index")
ax.set_ylabel("Y")

ax.set_title("my graph")
```

```julia
fig, axes = subplots(1, 2, figsize=(10, 3))
axes[1].plot(Y1)
axes[2].plot(Y2)
```

```julia
func(x) = x^2 # 関数定義はこれでも良い

xs = range(-3, 3, length=10) # length=10 -> 10等分の意味

fig, ax = subplots()

ax.plot(xs, func.(xs), "-")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("y=x^2")
```

```julia
func(x) = x^2 # 関数定義はこれでも良い

xs = range(-3, 3, length=100) # length=10 -> 10等分の意味

fig, ax = subplots()

ax.plot(xs, func.(xs), "-")

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("y=x^2")
```

```julia
r = 1.0
fx(θ) = r*(θ - sin(θ))
fy(θ) = r*(1 - cos(θ))
θs = range(-3pi, 3pi, length=100)

fig, ax = subplots()

ax.plot(fx.(θs), fy.(θs))

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("cycloid")
```

```julia
sig(x) = 1 / (1 + exp(-x))

xs = range(-5, 5, length=100) # length=10 -> 10等分の意味

fig, ax = subplots()

ax.plot(xs, sig.(xs))

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("sigmoid")
```

```julia
fz(x, y) = exp(-(2x^2+y^2+x*y))

xs = range(-1, 1, length=100)
ys = range(-2, 2, length=100)

fig, axes = subplots(1, 2, figsize=(8, 4))

cs = axes[1].contour(xs, ys, fz.(xs', ys))
axes[1].clabel(cs, inline=true)

cs = axes[2].contourf(xs, ys, fz.(xs', ys))
fig.colorbar(cs)

tight_layout()
```
