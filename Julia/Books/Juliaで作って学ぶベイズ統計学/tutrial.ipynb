```julia
using Plots
```

```julia
plot(sin)
```

```julia
plot(cos)
```

```julia
plot(
    x->x^3+x^2+x,       # 無名関数で描画する関数を決定
    xlabel="x",         # x軸凡例
    ylabel="y",         # y軸凡例
    label="y=x³+x²+x",  # グラフラベル
    xlims=(-3, 3),      # x軸範囲
    ylims=(-10, 10),    # y軸範囲
    xticks=-3:0.5:3,    # x軸目盛り表示刻み  
    title="Plot()",     # グラフタイトル
    linecolor=:orange,  # 色
    linewidth=3,        # 太さ
    linestyle=:dot,     # 描画スタイル
)
```

```julia
plt1 = plot(x->x)
plt2 = plot(x->x^2)
plt3 = plot(x->x^3)
 
plot(plt1, plt2, plt3, layout=(1, 3), size=(800, 400))
```

```julia
plot(sin, label="sin()")
plot!(cos, label="cos()")
plot!(x->0.1x^2, label="0.1x²", ylim=(-1, 1))
```

```julia
plt = plot(sin, label="sin()")
plot!(plt, cos, label="cos()")
plot!(plt, x->0.1x^2, label="0.1x²", ylim=(-1, 1))
```

```julia
# !は関数の引数を変更する際に使う
plot!(title="The comparison", xlim=(-4, 4))
```

```julia
p1 = bar(rand(50), title="bar")
p2 = pie(rand(8), title="pie")
p3 = histogram(randn(1000), title="histogram")
p4 = scatter(rand(20), rand(20), title="scatter")
 
x = y = -10:0.1:10
z = sin.(x) * cos.(y')  # "."を使った関数呼び出しは、配列に対して一括で関数処理を行うJuliaの記法
p5 = heatmap(z, title="heatmap")
 
p6 = surface(z, title="surface", color=palette(:ice))  # カラーパレットの変更
 
plot(p1, p2, p3, p4, p5, p6, layout=(3, 2), size=(1000, 800))
```
